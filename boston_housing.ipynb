# 爆速Boston housingインサイト？

## このノートブックで使用するライブラリ
- [pandas](https://pandas.pydata.org/)
- [scikit-learn](https://scikit-learn.org/stable/)

```
from sklearn.datasets import load_boston
import pandas as pd
```

```
boston = load_boston()
boston_df = pd.DataFrame(boston.data, columns=boston.feature_names)
boston_df['MEDV'] = boston.target
```

## データを見てみる


```
boston_df.head()
```

|カラム名|説明|
|----|----|
|CRIM	|犯罪発生率|
|ZN	|25,000平方フィート以上の住宅区画の割合|
|INDUS	|非小売業種の土地面積の割合|
|CHAS|	チャールズ川沿いかを表すダミー変数|
|NOX|	窒素酸化物の濃度|
|RM	|平均部屋数|

|PTRATIO|	生徒と教師の割合|
|B|	黒人の割合|
|LSTAT|	低所得者の割合|
|MEDV	|住宅価格の中央値（1,000単位）|

## データフレームの各カラムの相関を一括計算

相関ってなに？→[26-3. 相関係数 | 統計学の時間 | 統計WEB - BellCurve](https://bellcurve.jp/statistics/course/9591.html)

```
 corr = boston_df.corr()
```

```
import matplotlib.pyplot as plt

corr.style.format("{:.2}").background_gradient(cmap=plt.get_cmap('bwr'), axis=1)
```

## 住宅価格高いところって、リッチな人が多くて、治安が良さそう？

```
fig = go.Figure()
fig.add_trace(go.Scatter(x=boston_df["MEDV"].values,y=boston_df["CRIM"].values, 
                    mode='markers'))

fig.update_layout(
    title="犯罪発生率と住宅価格の中央値の関係",
    xaxis_title="住宅価格の中央値",
    yaxis_title="犯罪発生率",
    font=dict(
        family="Courier New, monospace",
        size=18,
        color="#7f7f7f"
    )
)

fig.show()
```

```
fig = go.Figure()
fig.add_trace(go.Scatter(x=boston_df["TAX"].values,y=boston_df["NOX"].values, 
                    mode='markers'))

fig.update_layout(
    title="",
    xaxis_title="10000ドルあたりの不動産税率",
    yaxis_title="窒素酸化物の濃度",
    font=dict(
        family="Courier New, monospace",
        size=18,
        color="#7f7f7f"
    )
)

fig.show()
```
